## Ініціазація індукції

```python
I = [P[i]*B[i, O[0]-1] for i in range(N)]
I
```

## Індукційний перехід

```python
from typing import List
def next(prev: List[float], t: int) -> List[float]:
    return [sum([prev[i]*A[i,j] for i in range(N)])*B[j, O[t]-1] for j in range(N)]

R = I
for i in range(len(O)-1):
    R = next(R, i+1)
    print(R)
```
