## Середнє

```python
def mean() -> np.array:
    return sum(data[i] for i in range(m))/m

M = mean()
M
```

## Коваріаційні матриці

```python
def matrix() -> np.array:
    return sum(
        np.array([
        (data[i, :, j]-M[:, j]).reshape(k,1)
            .dot((data[i, :, j]-M[:, j]).reshape(1,k))
                         for j in range(n)
        ])
               for i in range(m)
               )/(m-1)

CO = matrix()
CO[14] #t15
```
